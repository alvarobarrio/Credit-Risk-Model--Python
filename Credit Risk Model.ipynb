### 4. Discrete Variables: Automating Calculations

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[: ,[0, 1, 3]]                                   #specific columns
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']     #specific column names
    df['prop_n_good'] = df['n_obs'] / df['n_obs'].sum()           #proportion of obs from the total
    df['n_good'] = df['prop_good'] * df['n_obs']                  #number of good borrowers
    df['n_bad'] = (1- df['prop_good']) * df['n_obs']              #number of bad borrowers
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()         #proportion of good from total
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()            #proportion of bad from total
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])      #weight of evidence
    df = df.sort_values(['WoE'])                                  #sort by weight
    df = df.reset_index(drop = True)                              #reset the index
    df['diff_prop_good'] = df['prop_good'].diff().abs()           #differences in proportion of good 
    df['diff_WoE'] = df['WoE'].diff().abs()                       #and weight evidence
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE'] #Information Value
    df['IV'] = df['IV'].sum()
    return df
    
    
    
    
```

```python
df_temp = woe_discrete(df_inputs_prepr, 'grade', df_targets_prepr)
df_temp
```

```python
def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0): #rotation =  parameter that rotates the labels
    x = np.array(df_WoE.iloc[: , 0].apply(str)) #matplotlib deos not fare with dfs and strings
    y = df_WoE['WoE']
    plt.figure(figsize = (18, 6))
    plt.plot(x, y, marker = 'o', linestyle = '--', color = 'k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title(str('Weight of Evidence by ' + df_WoE.columns[0]))
    plt.xticks(rotation = rotation_of_x_axis_labels)
        
                 
                 
```

### 8. Continuous Variables Automating Calculations and Visualizing Results

```python
def woe_ordered_continuous(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[: ,[0, 1, 3]]                                   #specific columns
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']     #specific column names
    df['prop_n_good'] = df['n_obs'] / df['n_obs'].sum()           #proportion of obs from the total
    df['n_good'] = df['prop_good'] * df['n_obs']                  #number of good borrowers
    df['n_bad'] = (1- df['prop_good']) * df['n_obs']              #number of bad borrowers
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()         #proportion of good from total
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()            #proportion of bad from total
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])      #weight of evidence
    #we want to preserve their natural order
    #df = df.sort_values(['WoE'])                                  #sort by weight
    #df = df.reset_index(drop = True)                              #reset the index
    df['diff_prop_good'] = df['prop_good'].diff().abs()           #differences in proportion of good 
    df['diff_WoE'] = df['WoE'].diff().abs()                       #and weight evidence
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE'] #Information Value
    df['IV'] = df['IV'].sum()
    return df
```
